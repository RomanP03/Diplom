```python
import pandas as pd
import numpy as np

import os

import matplotlib.pyplot as plt
```

```python
home = os.getcwd()
```

```python
data = pd.read_csv(f'{home}/songs/Lyrics_LinkinPark_LVTTRHLN1V.csv')
data.sort_values(by=['Hapax legomena'])
```

```python
data['N_1/V'].mean()
```
