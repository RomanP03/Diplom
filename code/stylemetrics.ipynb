```python
import pandas as pd
import os
```

READ LV FILE

```python
lv_file = input("Enter LV_file name: ")
df = pd.read_csv(f'{lv_file}.csv')
df
```

**type-token ratio TTR = V/N**

```python
df['TTR'] = df['Кількість унікальних слів'] / df['Кількість слів']
df
```

**Слова, що трапляються тільки раз (hapax legomena)**

```python
df['Hapax legomena'] = ''
df
```

```python
n_folder_path = 'Lyrics_LinkinPark_ranks/'

# Проходимося по кожному файлу в папці
for filename in os.listdir(n_folder_path):

    # Перевіряємо, чи файл є csv
    if filename.endswith('.csv'):

        # Завантажуємо файл
        file_path = os.path.join(n_folder_path, filename)
        df_r = pd.read_csv(file_path)

        # Обчислюємо кількість hapax legomena
        hapax_count = (df_r['Count'] == 1).sum()

        # Знаходимо рядок у файлі l_v з назвою поточного файлу
        filename_without_suffix = filename.rstrip('_ranked')
        row_idx = df.index[df['Назва файлу'] == filename_without_suffix].tolist()[0]


        # Вносимо отримане значення у колонку "Hapax legomena"
        df.at[row_idx, 'Hapax legomena'] = hapax_count
df
```

**Частка слів, що трапляються тільки раз (hapax legomena) типу N_1/V, де N_1 — кількість унікальних слів, а V — розмір словника для кожного окремого треку.**
