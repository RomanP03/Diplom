```python
i = 0
data_inside = {}

for file in os.listdir(f'{directory}/{folder}'):
    vocab_counter = 0
    length_counter = 0
    vocab = []
    with open(f'{directory}/{folder}/{file}', 'r') as f:
        text = f.read().split(' ')
        for word in text:
            if not word in vocab:
                vocab.append(word)
    
    
    data_inside[file] = (len(text), len(vocab))
```

```python
data = pd.DataFrame.from_dict(data_inside, orient='index').reset_index()#, columns=['name', 'length', 'vocab'])
data = data.rename(columns={'index': 'name', 0: 'length', 1: 'vocab_size'})

data = data.sort_values(by=['length'], ignore_index=True)
data
#data.sort_values(by=['vocab_size'])
```

```python
fig, ax = plt.subplots(dpi=300)

ax.set(xscale='log', yscale='log', xlabel=r'$L$', ylabel=r'$V$')

ax.scatter(data['length'], data['vocab_size'], edgecolor='black', facecolor='white')

plt.savefig(f'{directory}/figures/unity-corpuse.png')
```

```python
alpha_1 = []
alpha_2 = []
gamma   = []

windows = [i for i in range(5, 300)]

for window in windows:
    step = window
    ls = data['length'].rolling(window=window, step=step).mean()
    vs = data['vocab_size'].rolling(window=window, step=step).mean()
    dv = data['vocab_size'].rolling(window=window, step=step).std()
    
    slope, intercept, r, p, se = stats.linregress(np.log(ls.dropna()), np.log(vs.dropna()))
    
    alpha_1.append(slope)
    
    slope, intercept, r, p, se = stats.linregress(np.log(ls.dropna()), np.log(dv.dropna()))
    
    alpha_2.append(slope)
    
    slope, intercept, r, p, se = stats.linregress(np.log(vs.dropna()), np.log(dv.dropna()))
    
    gamma.append(slope)
```

```python
fig, ax = plt.subplots(dpi=300)

ax.set(xlabel='window size', ylabel=r'$\gamma$')

ax.scatter(windows, alpha_1, color='black')

# plt.savefig(f'{directory}/figures/unity_corpus_window_size_alpha_1.png')
plt.savefig(f'{directory}/figures/songs_window_size_gamma.png')
```

```python
fig, ax = plt.subplots(dpi=300)

ax.set(xlabel='window size', ylabel=r'$\alpha_1$')

ax.scatter(windows, alpha_2, color='black')

# plt.savefig(f'{directory}/figures/unity_corpus_window_size_alpha_2.png')
plt.savefig(f'{directory}/figures/songs_window_size_alpha_1.png')
```

```python
fig, ax = plt.subplots(dpi=300)

ax.set(xlabel='window size', ylabel=r'$\alpha_2$')

ax.scatter(windows, gamma, color='black')

# plt.savefig(f'{directory}/figures/unity_corpus_window_size_alpha_1.png')
plt.savefig(f'{directory}/figures/unity_corpus_window_size_alpha_2.png')
```
