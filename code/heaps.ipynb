```python
# with open(f'{directory}/{folder}/{filename}', 'r') as file:
#         lines = file.read()

# text = lines.split(' ')

def get_vocab(text):
    
    unique = {}
    for word in text:
        if word in unique:
            unique[word] += 1
        else:
            unique[word] = 1
            
    return unique

def number_unique_words(vocab):
    return len(vocab)
```

```python
def heaps(x, D, gamma):
    return D*np.power(x, gamma)
```

```python
with open(f'{directory}/corpus/mergers/songs.txt', 'r') as file:
# with open(f'{directory}/processed/EditorGUI.txt', 'r') as file:
    lines = file.read()

text = lines.split(' ')

lengths = []
vocabs = []

step = 100

for i in range(100, len(text), step):
    start = time.perf_counter()
    subtext = text[:i]
    subtext_time = time.perf_counter()
    
#     print(f'subtext time {subtext_time - start}')
    
    lengths.append(len(subtext))
    length_time = time.perf_counter()      
    
#     print(f'length append time {length_time - subtext_time}')
    
    vocabs.append(len(get_vocab(subtext)))
    vocab_time = time.perf_counter()
    
#     print(f'vocab append time {vocab_time - length_time}')
    
    step += 50
    
    end = time.perf_counter()
    
    print(f'loop time {(end - start):.3}')
    
#print(vocabs)
# print(len(text))

# positions = []
# new_words = []
# vocab = []

# i = 1
# for word in text:
#     if word in vocab:
#         positions.append(i)
#         new_words.append(len(vocab))
#         i += 1
#     else:
#         positions.append(i)
#         vocab.append(word)
#         new_words.append(len(vocab))
#         i += 1
```

```python
vocab = pd.DataFrame.from_dict(unique, orient='index').reset_index()#, columns=['word', 'freq'])\
vocab = vocab.rename(columns={"index": "word", 0: "freq"})
vocab = vocab.sort_values(by='freq', ascending=False, ignore_index=True)
vocab.index = vocab.index + 1
vocab['norm_freq'] = vocab['freq'] / vocab['freq'].sum()
vocab
```

```python
def zipf_mandelbrot(x, A, alpha, r_0):
    return A / (r_0 + x)**alpha
```

```python
zm_model = Model(zipf_mandelbrot)

zm_params = zm_model.make_params(A=100, alpha=1.5, r_0=10)

zm_result = zm_model.fit(x=vocab.index, data=vocab['norm_freq'], params=zm_params)
```

```python
print(zm_result.fit_report())
```

```python
slope, intercept, r, p, se = stats.linregress(np.log(vocab.index), np.log(vocab['norm_freq']))

print(slope, r)
```

```python
fig, ax = plt.subplots(dpi=300)

ax.set(xscale='log', yscale='log', xlabel=r'$r$', ylabel=r'$f(r)$')

ax.scatter(vocab.index, vocab['norm_freq'], color='black', label='data')
ax.plot(vocab.index.to_numpy(), np.e**intercept * vocab.index.to_numpy()**slope, '--', color='red', label='Zipf')
ax.plot(np.asarray(vocab.index), np.asarray(zm_result.best_fit), '-.', color='blue', label='Zipf-Mandelbrot')

ax.legend()
```
